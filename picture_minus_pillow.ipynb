```python
from PIL import Image
import matplotlib.pyplot as plt
import PIL
import cv2
import numpy as np
```

```python
plt.subplot(121)
img1 = Image.open("images/input.png")
plt.imshow(img1, cmap='gray')

plt.subplot(122)
img2 = Image.open("images/input_rebuild_high200.png")
plt.imshow(img2, cmap="gray")
```

```python
img3 = PIL.ImageChops.difference(img1, img2)
plt.imshow(img3,cmap="gray")
img3.save("cut_output.jpg")
```
